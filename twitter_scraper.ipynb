```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
maxTweets = 1000000
tweets_list2 = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('tokenization AND investment  since:2009-10-31 until:2022-10-06').get_items()):
    if i>maxTweets:
        break
    if tweet.user.followersCount<5000:
        continue
    tweets_list2.append([tweet.id,tweet.date, tweet.content, tweet.likeCount, tweet.replyCount, tweet.user.username, tweet.user.followersCount, tweet.user.linkUrl, tweet.user.linkTcourl])

tweets_list2
```

```python
tweets_df = pd.DataFrame(tweets_list2, columns=['tweet_id', 'Datetime', 'tweet_content', 'like_count', 'reply_count', 'user_name', 'followers_count', 'user_linkUrl', 'user_linkTcourl'])
tweets_df.head()
print(tweets_df.shape)
len(tweets_df.user_name.unique())
```

```python
grouped_df = tweets_df.groupby('user_name')
```

```python
tweets_df[['user_name', 'followers_count', 'user_linkUrl', 'user_linkTcourl']].drop_duplicates().to_csv('Users.csv', index=False)
tweets_df.to_csv('tweets_csv')
```
